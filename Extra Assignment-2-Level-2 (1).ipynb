# print Even numbers in the list
# (20,10,16,19,25,1,276,188)

```python
Numbers = (20,10,16,19,25,1,276,188)
for num in Numbers:
    if num % 2==0:
        print(num)
```

# print Odd Numbers in the list
# (20,10,16,19,25,1,276,188)

```python
Numbers = (20,10,16,19,25,1,276,188)
for num in Numbers:
    if num % 2 != 0:
        print(num)
```
